```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bar_chart_race as bcr
from IPython.display import HTML
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Video
```

```python
covid_data_complete = pd.read_excel('/home/archman/Downloads/covid_india.xlsx', index_col=False)
covid_data_complete.head()
```

```python
covid_data = covid_data_complete.copy() #make a copy for analysis
covid_data.columns = ['Date', 'States', 'Cases'] #rename columns
covid_data.head(10)
```

```python
total_countries = covid_data['States'].nunique()
total_countries
```

```python
# set countries and date as index and find cases
# transpose the dataframe to have countries as columns and dates as rows
covid_data_by_date = covid_data.set_index(['States','Date']).unstack()['Cases'].T.reset_index()

covid_data_by_date = covid_data_by_date.set_index('Date') #make date as index - desired by barchartrace

covid_data_by_date = covid_data_by_date.fillna(0) #fill na with 0

covid_data_by_date
```

```python
#make the mp4 file with the BarChartRace and save it
df = covid_data_by_date
bcr.bar_chart_race(
    df=df,
    filename='India_Covid19_BarChartRace7.mp4',
    orientation='h',
    sort='desc',
    fixed_order=False,
    fixed_max=False,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .50, 'y': .20, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',
    period_length=1000,
    figsize=(8, 5),
    dpi=500,
    cmap='antique',
    title='COVID-19 Cases in India',
    title_size=10,
    bar_label_size=7,
    tick_label_size=6,
    shared_fontdict={'family': 'Arial', 'weight': 'bold',
                                    'color': 'rebeccapurple'},
    scale='linear',
    writer=None,
    fig=None,
    #bar_kwargs={'alpha': .2, 'ec': 'black', 'lw': 3},
    filter_column_colors=True) 
```
